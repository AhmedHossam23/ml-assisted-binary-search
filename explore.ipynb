```python
import pandas as pd
```

```python
df1 = pd.read_pickle(
    "data/data__samples_100000__features_100__n_informative_50__min_max_depth_1__max_max_depth_40.pkl")
df2 = pd.read_pickle(
    "data/data__samples_100000__features_100__n_informative_55__min_max_depth_1__max_max_depth_40.pkl")
```

```python
x = df1.T.join(df2.T, lsuffix="_50", rsuffix="_55")
```

```python
preds_50_informative = x["preds_50"]
preds_55_informative = x["preds_55"]
commits_50_informative = x["commits_50"]
```

```python
preds_50_informative

import matplotlib.pyplot as plt


# preds_50_informative is a series, each entry is a list of predictions for a given sequence of commits
# preds_50_informative[0] is a list of predictions for the first sequence of commits
# preds_50_informative[0][0] is the prediction for the first commit in the first sequence of commits

# I want to plot a KDE of the predictions for the first commit in the first sequence of commits
plt.figure()
plt.hist(preds_50_informative.iloc[0], bins=100, alpha=0.5, label="50 informative features")
plt.hist(preds_55_informative.iloc[0], bins=100,
         alpha=0.5, label="55 informative features")
plt.legend()
plt.show()
```
